## Data Cleaning:

### Handling Missing Values:

```python
missing_values = amazon_df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

### Checking and handling duplicates:

```python
total_duplicates = amazon_df.duplicated().sum()

print(f'Total number of duplicates considering all columns: {total_duplicates}')
```

## Changing Data Types:

### Converting 'discounted_price' and 'actual_price' columns to float:

```python
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace("₹",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace(",",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype('float64')

amazon_df['actual_price'] = amazon_df['actual_price'].str.replace("₹",'')
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace(",",'')
amazon_df['actual_price'] = amazon_df['actual_price'].astype('float64')
```

### Converting 'discount_percentage' to a float and adjusting values:

```python
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%','').astype('float64')

amazon_df['discount_percentage'] = amazon_df['discount_percentage'] / 100

amazon_df['discount_percentage']
```

### Top 10 Customers

```python
top_10_customers = amazon_df['user_name'].value_counts().nlargest(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_customers.values, y=top_10_customers.index, palette='viridis', hue=top_10_customers.index, legend=False)
plt.title('Top 10 Customers')
plt.xlabel('Number of purchases')
plt.ylabel('Customer Name')
plt.show()

print ("Top 10 Customers:")
print(top_10_customers)
```

### Top 10 Categories

```python
top_10_categories = amazon_df['category'].value_counts().nlargest(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_categories.values, y=top_10_categories.index, palette='Set2', hue=top_10_categories.index, legend=False)
plt.title('Top 10 Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()

print ("Top 10 Categories:")
print(top_10_categories)
```

### Top 10 Products

```python
top_10_products = amazon_df['product_name'].value_counts().nlargest(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_products.values, y=top_10_products.index, palette='mako', hue=top_10_products.index, legend=False)
plt.title('Top 10 Products')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Name')
plt.show()

print ("Top 10 Products:")
print(top_10_products)
```
