## Data Cleaning:

### Handling Missing Values:

```python
missing_values = amazon_df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

### Checking and handling duplicates:

```python
total_duplicates = amazon_df.duplicated().sum()

print(f'Total number of duplicates considering all columns: {total_duplicates}')
```

## Changing Data Types:

### Converting 'discounted_price' and 'actual_price' columns to float:

```python
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace("₹",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace(",",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype('float64')

amazon_df['actual_price'] = amazon_df['actual_price'].str.replace("₹",'')
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace(",",'')
amazon_df['actual_price'] = amazon_df['actual_price'].astype('float64')
```

### Converting 'discount_percentage' to a float and adjusting values:

```python
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%','').astype('float64')

amazon_df['discount_percentage'] = amazon_df['discount_percentage'] / 100

amazon_df['discount_percentage']
```
